```python
import logging as lg
lg.basicConfig(filename='word_operation.log', level=lg.INFO,  format='%(asctime)s %(levelname)s %(message)s')
```

```python
from bagOfWords.wordOps import WordOperation
```

```python
try:
    obj=WordOperation(folderpath='words_file')
    lg.info('initialized process')
except Exception as e:
    lg.exception(e)
    print(e)
```

```python
try:
    word_count_list=obj.getWordAndCount(show_words=100)
    lg.info('getting word_count_list')
    print(word_count_list)
except Exception as e:
    lg.exception(e)
    print(e)
```

```python
try:
    alphabet_count_list=obj.getAlphabetAndCount()
    lg.info('got count of words starting with each alphabet')
    print(alphabet_count_list)
except Exception as e:
    lg.exception(e)
    print(e)
```

```python
try:
    msg=obj.filterData()
    lg.info('filtered data inside files and saved filtered files into a new folder')
    print(msg)
except Exception as e:
    lg.exception(e)
    print(e)
```

```python
try:
    msg=obj.insertIntoSqliteDb('words.db')
    lg.info('Data inserted into sqlite db')
    print(msg)
except Exception as e:
    lg.exception(e)
    print(e)
```

```python
try:
    data=obj.showData('words.db')
    for i in data:
        print(i)
    lg.info('all data in db showed')
except Exception as e:
    lg.exception(e)
    print(e)
```
